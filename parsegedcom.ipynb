```python

TAGS = {#"TAG":LEVEL
    "NOTE": 0,
    "TRLR": 0,
    "HEAD": 0,
    "DATE": 2,
    "DIV": 1,
    "CHIL": 1,
    "WIFE": 1,
    "HUSB": 1,
    "MARR": 1,
    "FAM": 0,
    "FAMS": 1,
    "FAMC": 1,
    "DEAT": 1,
    "BIRT": 1,
    "SEX": 1,
    "NAME": 1,
    "INDI": 0,
}
id = ""
def valid_tag(level, tag):
    return "Y" if (tag in TAGS and TAGS[tag] == level) else "N"
with open('GEDCOM/Simple.ged', "r+") as file:
    for line in file:
        section = line.rstrip().split(" ")
        level, tag, *arguments = section
        if arguments and arguments[0] in ("INDI","FAM"):#INDI & FAM should be in 2nd position
            new_tag = arguments[0]
            arguments[0] = tag
            print("-->" + line, end='') #Prints "--> <input line>"
            print(f"<-- {level}|{new_tag}|{valid_tag(int(level), new_tag)}|{' '.join(arguments)}") #Prints "<-- <level>|<tag>|<valid?> : Y or N|<arguments>"
        else:
            print("-->" + line, end='') #Prints "--> <input line>"
            print(f"<-- {level}|{tag}|{valid_tag(int(level), tag)}|{' '.join(arguments)}") #Prints "<-- <level>|<tag>|<valid?> : Y or N|<arguments>"
```
